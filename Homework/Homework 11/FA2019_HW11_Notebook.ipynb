```python
import numpy as np
import matplotlib.pyplot as plt
```

### Problem 2 Data

```python
nodes = 19
x = np.linspace(-1,2.1,nodes)
xi = np.linspace(-0.8,2.05,nodes)
y = np.array([2.71148102, 2.57622207, 2.29871322, 1.94086829, 1.56817399, 1.22977295, 0.95124924, 0.7382513 , 0.58460275, 0.47972235, 0.41332379, 0.37762092, 0.36803634, 0.38329695, 0.4254211, 0.49973115, 0.61467762, 0.78088121, 1.00844269])
phi= np.vander(x,N=nodes,increasing=True)
c = np.linalg.solve(phi,y)
yi=np.dot(phi,c)
plt.plot(x,np.dot(phi,c),'ro',x,y,'b.',xi,yi,'g.')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Collocation and Interpolation');
for x in range(0,len(xi)):
    print('The interpolated point is ({:1.2f},{:1.2f})'.format(xi[x],yi[x]))
#print(xi,yi)
#print(yi)
```

### Problem 3

```python
N = 20 # You may change N
x = np.linspace(-np.pi,np.pi,N)
x = x[:-1] # Evaluations at first and last endpoint are added together due to periodicity
# k = -Nk/2+1 ... Nk/2
# Frequency grid
Nk = 8
k = np.arange(-Nk/2+1,Nk/2+1) # Recall arange leaves out last value
# Outer Product for Fitting
kx = np.outer(k,x)
S = np.sin(np.outer(k,x))

f = (x<0)*(x-np.pi/2) + (x>=0)*(x+np.pi/2) # Here is one way to construct f
b = S.dot(f)/(N-1)

xi = np.linspace(-2.8,2.8)
Ni = len(xi)
kxi = np.outer(k,xi) # Outer Product for Interpolating
Si = np.sin(kxi)
Fsi = b.dot(Si) # Sine (odd part) Interpolant

plt.plot(x,f,'bo',xi,Fsi,'r.') # Interpolant
plt.axis([-np.pi,np.pi,np.min(f)-0.25,np.max(f)+0.25])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fourier Series Expansion (Interpolation)')
plt.legend(['Data','Fourier Sine Series']);
plt.plot(x,f,'r.');
```

### Problem 4

```python
N = 26 # You may change N
L = np.pi
# Spatial grid
x = np.linspace(-L,L,N)
x = x[:-1] # Evaluations at first and last endpoint are added together due to periodicity
# k = -Nk/2+1 ... Nk/2
# Frequency grid
Nk = 8
k = np.arange(-Nk/2+1,Nk/2+1) # Recall arange leaves out last value
# Outer Product for Fitting
kx = np.outer(k,x)
C = np.cos(np.outer(k,x))

g = (x<0)*(x+1) + (x>=0)*(1-x) # Here is one way to construct f
a = C.dot(g)/(N-1)

xi = np.linspace(-2.8,2.8)
Ni = len(xi)
kxi = np.outer(k,xi) # Outer Product for Interpolating
Ci = np.cos(kxi)
Gci = a.dot(Ci) # Cosine (even part) Interpolant

plt.plot(x,g,'bo',xi,Gci,'r.') # Interpolant
plt.axis([-L,L,np.min(g)-0.25,np.max(g)+0.25])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fourier Series Expansion (Interpolation)')
plt.legend(['Data','Fourier Cosine Series']);
```
