```python
#This was used to get accurate estimate for number of terms for question 4
def W(x,nterms) :
    p = 0
    for k in range (nterms) :
        p += (-1)**k*(x**(2*k+1)/(2*k+1))
    return p
```

```python
x = np.linspace(0,1)
y = W(x,500)
d = np.arange(1,2)
p = np.prod(d)
plt.plot(x,y,'r');
```

```python
#This is Question 3
def Z(x,nterms):
    q = 0
    for k in range (nterms) :
        q += (-1)**k*(((np.pi/2)**(2*k+1))*((x)**(4*k+2)))/(np.math.factorial(2*k+1))
    return q
```

```python
x= np.linspace(0,2) 
y = Z(x,10) 
plt.plot(x,y,'r');
```

```python
Z(1,7)
```
